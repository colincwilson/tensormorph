One-shot learning / initialization of (a sequence of) cogrammars with program synthesis

- given (input, output) pair

- sample a sequence of cogrammars and the type of each one (e.g., reduplication vs. non-redup)

- for each cogrammar
    - sample pivot parameters, truncation parameters  
     
    - sample affix length but not affix form: instead represent affix as a sequence of the special filler ☐ (e.g., representing uniform distribution over segments) [this is similar to an 'unknown constant', denoted '??', in Solar-Lezama's Sketch, which is instantiated in context]


- run network in forward direction from input and calculate decoding probability of output; train affix form with gradient descent

This method of one-shot learning / initialization can be done entirely within the network using hard or soft sampling (no external symbolic supervision, cf. string alignment initialization above). Importantly, we can score any proposed analysis against the example or batch of examples (e.g., compute its likelihood), and we could even use the gradient to identify 'problem areas' in the output to guide reanalysis. We can also include regularization/prior terms to prefer concise analyses

**Stochastic search**: we can use iterative refinement (as in Gibbs sampling, evolutionary search) to resample while preserving good substructure in the search for a high-likelihood initialization

  - need to allow for birth/death moves for cogrammars in (at least) linear sequence from input to output  

  - Metropolis-Hastings for variable number of components (RJMCMC: e.g., Richardson & Green 1997; Roodaki, Bect & Fleury 2012; Lopes 'A Note on RJMCMC' explicitly gives cases in which Jacobian is one). cf. Alur et al. (2013) simply fix a size $n$ for a model and search in models of different sizes without 'jumping' from one dimensionality to another. See also the comparison of RJMCMC and non-parametric samplers in Wood, Griffiths & Ghahramani (2012).
  
    - **transdimensional** (unbounded set of models of different dimensionality) vs. **non-parametric** (single model with unbounded dimensionality)

```python
import torch
import edlib
nrole = 20

def fast_analysis(stem_str='>tristi<', output_str='>trumisti<'):
    # 1. Align stem and output strings xxx not space delimited
    result = edlib.align(output_str, stem_str, task='path')
    nice = edlib.getNiceAlignment(result, output_str, stem_str)
    aligned = [x for x in zip(nice['matched_aligned'],
                              nice['query_aligned'],
                              nice['target_aligned'])]
    nice = '\n'.join(nice.values())
    print(result['cigar'])
    print(nice)
    
    #2. Extract affixal, stem_pivot, affix_pivot, stem_copy, affix_copy, etc.
    affix = []
    stem_pivot = torch.zeros(nrole)
    affix_pivot = torch.zeros(nrole)
    stem_copy = torch.zeros(nrole)
    affix_copy = torch.zeros(nrole)
    stem_indx = 0
    affix_indx = 0
    n = len(aligned)
    for i in range(0, n):
        a = aligned[i]
        a_prev = aligned[i-1] if i>0 else [None]*3
        a_next = aligned[i+1] if i<(n-1) else [None]*3
        # read from stem (possibly with substitution)
        if (a[0] == '|') or (a[0] == '.'):
            if (a_next[0] == '-') and (a_next[1] != '-'):
                stem_pivot[stem_indx] = 1.0
            stem_copy[stem_indx] = 1.0
            stem_indx = stem_indx + 1
        # delete from stem
        if (a[0] == '-') and (a[1] == '-'):
            stem_copy[stem_indx] = 0.0
            if (a_next[0] == '-') and (a_next[1] != '-'):
                stem_pivot[stem_indx] = 1.0
            stem_indx = stem_indx + 1
        # insert affix
        if (a[0] == '-') and (a[2] == '-'):
            affix.append(a[1])
            affix_copy[affix_indx] = 1.0
            if a_next[0] != '-':
                affix_pivot[affix_indx] = 1.0
            affix_indx = affix_indx + 1
    affix_str = ' '.join(affix) # xxx append end delimiter?
    val = {'stem_pivot': stem_pivot,
           'stem_copy': stem_copy,
           'affix_str': affix_str,
           'affix_pivot': affix_pivot,
           'affix_copy': affix_copy}
    return val

#val = fast_analysis(stem_str='>tristi<', output_str='>trumisti<')
#val = fast_analysis(stem_str='>wofe<', output_str='>wowofe<')
val = fast_analysis(stem_str='>ulampor<', output_str='>ulalampor<')
# xxx need to implement affix_form fill-in for reduplication: recursively apply alignment
# to 'affix' identified by initial alignment
for x in ['stem_pivot', 'stem_copy', 'affix_str', 
          'affix_pivot', 'affix_copy']:
    print(f'{x}: {val[x]}')
```

06/17/20

- Pivoter and truncater identify a fixed set of positions/segments in an input form and therefore _need not be differentiable_ (though selection from possible pivots/segments by softmax must be differentiable)  

    - e.g., can locate all pivot positions / truncation segments with non-differentiable impulse / rectangular functions and then detach()  
  
  
- Two ways of implementing selection among pivot points (same applies for truncation), corresponding roughly to soft vs. hard attention:

$$\texttt{output} = \textbf{combine}(\texttt{stem}, \texttt{affix}, \texttt{pivot} = \sum_{i} p_i \texttt{pivot}_i)$$

$$\texttt{output} = \sum_i p_i \textbf{combine}(\texttt{stem}, \texttt{affix}, \texttt{pivot}_i)$$

The first 'soft' way will produce partial-pivoting blends when the distribution over pivots is not peaked, while the second 'hard' way always blends the outputs of exact pivoting. Could also make affix pivot-dependent in 'hard' method, replacing $\texttt{affix}$ with $\texttt{affix}_i$

06/06/20

- Consider **hard attention** (aka alignment) model of Wu, Shapiro & Cotterell (2018)

- Can use hard attention during (final) decoding by creating a probabilistic FSA that represents a mixture of softmax distributions at each **decoding step** (not necessarily equal to each **output position**)

- Probability of output symbols at decoding step $t$ is given by

$$p_t(y) = \sum_{m \in \{stem, affix\}} \sum_{i=0}^{nrole-1} p_t(m) \ p_t(i | m) \ p(y | x_{mi})$$

where the probability of $\epsilon$ (i.e., deletion) is implicitly included

06/06/20

**Inferring latent alignments**

- A submodel can infer latent alignments **(stem, affix) -> output** by inferring **(copy_stem, copy_affix, pivot, unpivot)** parameters, each [nbatch x nrole], and **Affix** parameters [dfill x nrole]. This converges quickly to reasonable alignments that reconstruct the output from the input ('synthesis-by-anaysis').

- Alternative version of aligner would directly learn softmax parameters over morphs (stem, affix) and positions within morphs (0, ..., nrole-1) at each decoding step (not necessarily equal to each output step because of possibility of decoding epsilons). This version has many more parameters and unsurprisingly is much slower to converge.

see Wu & Cotterell, Raffel et al. (2017), Aharoni & Goldberg (2016), Makarov & Clematide (2018ab) on hard monotonic alignments

05/23/20

Transforming distance to similarity  

- Standard transformations are:

$$d(x_1, x_2) \to \frac{1}{1 + d(x_1, x_2)}$$

$$d(x_1, x_2) \to \exp(-d(x_1,x_2))$$

- For Euclidean distance in particular, the following identities:

$$||x_1 - x_2||^2 = (x_1 - x_2)^T (x_1 - x_2) = ||x_1|| + ||x_2|| - 2 x_1^T x_2$$

$$cos(\theta) = \frac{x_1^T x_2}{||x_1|| ||x_2||}$$

when $||x_1|| = ||x_2|| = 1$, these simplify to:

$$||x_1 - x_2||^2 = 2(1 - x_1^T x_2)$$

ref: https://en.wikipedia.org/wiki/Cosine_similarity

05/13/20

* Problem  
  - Consistent local minimum on Chamorro -um- infixation pattern: pivot immediately after initial delimiter, learn gradient three segment affix, fail on >= 40% of the data  
  - Best guess at internal analysis is as follows:  
    ex. input > p a d a < mapped to output > (p)+eps (p)+u (p)+m a d a < decoded as > p u m a d a <  
    where (x) is weakly-attended x  
  
  
* Diagnosis  
  - For affixes located near the beginning of the output, misalignment early on has cascading effects on later alignments during decoding (e.g., target talakhidzuŋ -> t-um-alakhidzuŋ aligned with incorrect t-u-alakhidzuŋ vs. um-talakhidzuŋ)  
  - Preliminary simulations with OpenNMT-py suggest that generic-encoder decoder networks might solve this problem by repeating segments in the affixation zone (or at least repeating focus of attention, talakhidzuŋ -> ttaalakhidzuŋ, then learning to overwrite segments with fixed affix), alternatively by strongly diverting attention away from segments on the one-to-one path while writing out affix segments  
  
  
* Solutions
  - Learning the affix is rapid (~ 10 epochs of 12 examples each) when the model is provided with the correct fixed pivoter  
  - x Tried replacing pivot finder with generic biLSTM with sigmoid or softmax output --- same local minimum  
  - Tried micro-ensembling with two fixed hard-coded pivoters, before_first_vowel and after_initial_delim --- solves the problem fairly reliably  
  - Reimplemented pivoter with restriction that only one positive feature value can determine pivot location (begin/end/C/V), using softmax and hardtanh judiciously. Works fairly consistently but there is still a strong local minimum in which the infix is mislearned as a prefix. Presumably due to cascading effects in left-to-right decoding.
  
* Solved by reverting pivot finding to LSTM/GRU recurrent unit. Lesson learned
